```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
from datetime import datetime

import statsmodels.api as sm

## prepare the data. 
## the following code was adapted from milestone 1 notebook

def prepare_data():
    df = pd.read_csv(os.path.join('.', 'dataset', 'energey_temperature_data.csv'))
    
    df['DATE'] = pd.to_datetime(df['DATE'])
    df.set_index('DATE', inplace=True)
    df.columns = ['mw', 'temp', 'hour_of_day', 'day_of_week', 'month', 'year']
    
    train_set = df[:'2019-12-31']
    test_set = df['2019-12-31':]
    
    return train_set, test_set

def preprocess(df):
    df['temp'] = (df['temp'] - np.mean(df['temp']))/np.std(df['temp'])
    df['temp_squared'] = df['temp'] ** 2
    return df

(train, test) = prepare_data()
train_df = preprocess(train.copy())
train_df.head(10)
```

### Save the residuals from your final linear regression model on your training data set from the previous section as their own variable.

```python
formula_expr2 = 'mw ~ C(hour_of_day) + C(day_of_week) + C(month) + temp * C(hour_of_day) + ' + \
                'temp_squared * C(hour_of_day) + temp * C(month) + temp_squared * C(month) + C(day_of_week) * C(hour_of_day)'
model5 = sm.OLS.from_formula(formula_expr2, data=train_df).fit()
residual = model5.resid
```

### Visualize your residuals across time in your training dataset.

```python
fig, ax = plt.subplots(figsize=(20, 5))

ax.plot(residual)
ax.set_ylabel('Residuals')
ax.set_xlabel('Date (hourly)')
ax.set_title("Residuals from the fitted model using OLS")

fig.autofmt_xdate()
plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(20, 5))

ax.plot(residual.loc['2019-01-01':'2019-01-07'])
ax.set_ylabel('Residuals')
ax.set_xlabel('Date (hourly)')
ax.set_title("Residuals from the fitted model using OLS for date range '2019-01-01' to '2019-01-07']")

fig.autofmt_xdate()
plt.tight_layout()
```
