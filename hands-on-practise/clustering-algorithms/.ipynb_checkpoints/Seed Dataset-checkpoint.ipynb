# Seed dataset classification
[See document of the dataset here](http://archive.ics.uci.edu/ml/datasets/seeds)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from urllib.request import urlopen

from sklearn.cluster import Birch, DBSCAN, KMeans
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

## Data extraction

```python
data_url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/00236/seeds_dataset.txt'
column_names = ['area_a', 'perimeter', 'compactness', 'kernel_length', 'kernel_width', 'asymmetry_coefficient', 'kernel_groove_length', 'class']
data = []
with urlopen(data_url) as remote_data:
    raw_data = remote_data.readlines() # data not much, we can read all at once
    for line in raw_data:
        parts = line.decode("utf-8") .replace('\\n', '').rstrip().split('\t')
        if len(parts) > 8:
            parts = list(filter(lambda x : len(x) > 0, parts))
        
        data.append(parts)
        
data = pd.DataFrame(np.array(data, dtype=np.float32), columns=column_names)
data.info()
```

```python
print('Any nulls: ' ,data.isnull().values.any())
print('Any NAs: ' ,data.isna().values.any())
```

# Dimensionality Reduction with PCA

```python
def get_principal_component_n(threshold=0.95):
    pca = PCA()
    pca.fit(X_scaled)
    explained_variance = pca.explained_variance_ratio_.cumsum()
    for i in enumerate(explained_variance):
        if i[1] >= threshold:
            return i[0] + 1
        
n = get_principal_component_n(0.96)
pca = PCA(n_components=n)
X_scaled_reduced = pd.DataFrame(pca.fit_transform(X_scaled), columns=[f'pc_{i+1}' for i in range(n)] )
X_scaled_reduced.head()
```

# Clustering with [BIRCH](https://scikit-learn.org/stable/modules/clustering.html#birch)
It is a memory-efficient, online-learning algorithm provided as an alternative to MiniBatchKMeans. It constructs a tree data structure with the cluster centroids being read off the leaf. These can be either the final cluster centroids or can be provided as input to another clustering algorithm such as AgglomerativeClustering.

```python
birch = Birch(n_clusters=3)
birch_res = birch.fit_transform(X_scaled)
# birch.labels_

fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(20,6))

ax[0].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=kmeans.labels_)
ax[0].set_xlabel('kernel_width')
ax[0].set_ylabel('asymmetry_coefficient')
ax[0].set_title('Cluster with KMeans')

ax[1].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=birch.labels_)
ax[1].set_xlabel('kernel_width')
ax[1].set_ylabel('asymmetry_coefficient')
ax[1].set_title('Cluster with Birch')

plt.show()
```

```python
birch_pca = Birch(n_clusters=3)
birch_pca_res = birch_pca.fit_transform(X_scaled_reduced)
# birch.labels_

fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(20,6))

ax[0].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=kmeans.labels_)
ax[0].set_xlabel('kernel_width')
ax[0].set_ylabel('asymmetry_coefficient')
ax[0].set_title('Cluster with KMeans + PCA')

ax[1].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=birch_pca.labels_)
ax[1].set_xlabel('kernel_width')
ax[1].set_ylabel('asymmetry_coefficient')
ax[1].set_title('Cluster with Birch + PCA')

plt.show()
```

# Clustering with [DBSCAN](https://scikit-learn.org/stable/modules/clustering.html#dbscan)
DBSCAN - Density-Based Spatial Clustering of Applications with Noise. Finds core samples of high density and expands clusters from them. Good for data which contains clusters of similar density.

```python
dbscan = DBSCAN(eps=0.9, min_samples=10)
dbscan.fit(X_scaled)
# print(dbscan.labels_)
# print(dbscan.components_)

fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(20,6))

ax[0].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=kmeans.labels_)
ax[0].set_xlabel('kernel_width')
ax[0].set_ylabel('asymmetry_coefficient')
ax[0].set_title('Cluster with KMeans')

ax[1].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=dbscan.labels_)
ax[1].set_xlabel('kernel_width')
ax[1].set_ylabel('asymmetry_coefficient')
ax[1].set_title('Cluster with DBSCAN')

plt.show()
```

```python
dbscan_pca = DBSCAN(eps=0.9, min_samples=10)
dbscan_pca.fit(X_scaled_reduced)
# print(dbscan_pca.labels_)
# print(dbscan.components_)

fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(20,6))

ax[0].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=kmean_pca.labels_)
ax[0].set_xlabel('kernel_width')
ax[0].set_ylabel('asymmetry_coefficient')
ax[0].set_title('Cluster with KMeans + PCA')

ax[1].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=dbscan_pca.labels_)
ax[1].set_xlabel('kernel_width')
ax[1].set_ylabel('asymmetry_coefficient')
ax[1].set_title('Cluster with DBSCAN + PCA')

plt.show()
```

##### DBSCAN is a wrong tool for this kind of dataset. # Clustering with [Gaussian mixtures](https://scikit-learn.org/stable/modules/mixture.html#mixture)

```python
gmm = GaussianMixture(n_components=3, covariance_type="full")
gmm_labels = gmm.fit_predict(X_scaled)

# plot compare
fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(20,6))

ax[0].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=kmeans.labels_)
ax[0].set_xlabel('kernel_width')
ax[0].set_ylabel('asymmetry_coefficient')
ax[0].set_title('Cluster with KMeans')

ax[1].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=gmm_labels)
ax[1].set_xlabel('kernel_width')
ax[1].set_ylabel('asymmetry_coefficient')
ax[1].set_title('Cluster with GaussianMixture')

plt.show()
```

```python
gmm_pca_labels = gmm.fit_predict(X_scaled_reduced)

# plot compare
fig, ax = plt.subplots(ncols=2, sharey=True, figsize=(20,6))

ax[0].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=kmeans.labels_)
ax[0].set_xlabel('kernel_width')
ax[0].set_ylabel('asymmetry_coefficient')
ax[0].set_title('Cluster with KMeans')

ax[1].scatter(data['kernel_width'], data['asymmetry_coefficient'], c=gmm_pca_labels)
ax[1].set_xlabel('kernel_width')
ax[1].set_ylabel('asymmetry_coefficient')
ax[1].set_title('Cluster with GaussianMixture + PCA')

plt.show()
```
