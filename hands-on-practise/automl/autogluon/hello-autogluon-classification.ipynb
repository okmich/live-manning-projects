# Hello Autogluon
An example from quick start guide on https://auto.gluon.ai/stable/tutorials/tabular/tabular-quick-start.html.

```python
from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.metrics import confusion_matrix
import os

model_save_path = os.path.join(os.environ['HOME'], 'autogluon_models', 'hello-autogluon')
```

```python
data_url = 'https://raw.githubusercontent.com/mli/ag-docs/main/knot_theory/'
train_data = TabularDataset(f'{data_url}train.csv')
train_data.head()
```

```python
#target variable is "signature"
label = 'signature'
train_data[label].describe()
```

## Training

```python
predictor = TabularPredictor(label=label, path=model_save_path)
predictor.fit(train_data);
```

## Prediction

```python
test_data = TabularDataset(f'{data_url}test.csv')

y_pred = predictor.predict(test_data.drop(columns=[label]))
y_pred.head()
```

## Evaluation

```python
predictor.leaderboard(test_data, silent=True)
```

```python
predictor.evaluate(test_data, silent=True)
```

```python
predictor.evaluate(test_data, model='LightGBM', silent=True)
```

```python
predictor.evaluate(test_data, model='XGBoost', silent=True)
```

```python
confusion_matrix(y_true=test_data[label], y_pred=y_pred)
```

```python
predictor.fit_summary(verbosity=3, show_plot=True)
```

## Load all models from model_save_path

```python
reloaded_models = TabularPredictor.load(model_save_path)
```

```python
reloaded_models.leaderboard(silent=False);
```

## Prepare for deployment

```python
deploy_location = os.path.join(os.environ['HOME'], 'autogluon_models', 'hello-autogluon-deployment-location')
predictor.clone_for_deployment(path=deploy_location);
```
