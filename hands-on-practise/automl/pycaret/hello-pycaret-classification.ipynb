```python
# load sample dataset
from pycaret.classification import ClassificationExperiment
from pycaret.datasets import get_data

import os

data = get_data('diabetes')
```

```python
s = ClassificationExperiment()
s.setup(data, target = 'Class variable', session_id = 123)
```

```python
best = s.compare_models()
```

```python
log_regress = s.create_model('lr')
```

```python
s.evaluate_model(best)
```

```python
s.plot_model(best, plot = 'confusion_matrix') # auc
```

```python
# predict 
s.predict_model(best)
```

```python
predictions = s.predict_model(best, data=data, raw_score=True)
predictions.head()
```

```python
# save model
save_path = os.path.join(os.getenv('HOME'), 'models', 'pycaret', 'hello_classification')
if not os.path.exists(save_path):
    os.makedirs(save_path)
s.save_model(best, save_path)
```

```python
# load the model backup
model = s.load_model(save_path)
model
```
