```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from typing import Callable, Tuple

from sklearn.metrics import mean_squared_error

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_process import ArmaProcess
```

### For this exercise, generate 10,000 samples of a four order moving average process.

```python
np.random.seed(42)

ma2 = np.array([1, 0.9, 0.3, -0.5, 0.03])
ar2 = np.array([1, 0, 0])

data_process = ArmaProcess(ar2, ma2).generate_sample(nsample=10000)
```

### Plot your simulated moving average.

```python
plt.figure(figsize=(20,5))
plt.plot(data_process);
plt.title('simulated moving average.')
plt.show();
```

### Plot acf and pacf

```python
plot_acf(data_process, title='ACF plot for the simulated moving average process')
plt.show() 
```

```python
plot_pacf(data_process, title='PACF plot for the simulated moving average process', method='ywm')
plt.show() 
```
