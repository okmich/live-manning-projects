```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from typing import Callable, Tuple

from sklearn.metrics import mean_squared_error

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.tsa.statespace.sarimax import SARIMAX
from statsmodels.tsa.stattools import adfuller
```

### For this exercise, generate 1,000 samples.

```python
np.random.seed(42)

ma2 = np.array([1, 0.9, 0.3])
ar2 = np.array([1, 0, 0])

data_process = ArmaProcess(ar2, ma2).generate_sample(nsample=1000)
```

### Plot your simulated moving average.

```python
plt.figure(figsize=(20,5))
plt.plot(data_process);
plt.title('simulated moving average.')
plt.show();
```

### 3 Run the ADF test, and check if the process is stationary.

```python
adf_test = adfuller(data_process, maxlag=30)
adf_message = 'Stationary' if adf_test[1] < 0.05 else 'Not stationary'
print(adf_test)
print(f'The result is {adf_message}')
```

### 4 Plot the ACF, and see if there are significant coefficients after lag 2.

```python
plot_acf(data_process, title='ACF plot for the simulated moving average process')
plt.show() # acf plot show significance only on first 3 lags
```

## # 5 Separate your simulated series into train and test sets. Take the first 800 timesteps for the train set, and assign the rest to the test set.

```python
horizon_steps = 200
train = data_process[:-horizon_steps]
test = data_process[-horizon_steps:]
```

### Use ARIMA with straight prediction/forecast

```python
arima = ARIMA(train, order=(0,0,3))
arima_fit = arima.fit()
arima_fittedvalues = arima_fit.fittedvalues
prediction_result = arima_fit.get_forecast(steps=horizon_steps)
```

```python
plt.figure(figsize=(20,5))
plt.plot(test, label='actual')
plt.plot(prediction_result.predicted_mean, label='prediction')
plt.title('simulated moving average.')
plt.legend()
plt.show();
```

### Use ARIMA with rolling prediction/forecast

```python

def rolling_forecast(arr: np.ndarray, train_len: int, horizon: int, window: int, method: str, order: Tuple[int, int, int]) -> list:
    total_len = train_len + horizon
    if method == 'mean':
        pred_mean = []
        for i in range(train_len, total_len, window):
            mean = np.mean(arr[:i])
            pred_mean.extend(mean for _ in range(window))
        return pred_mean
    elif method == 'last':
        pred_last_value = []
        for i in range(train_len, total_len, window):
            last_value = arr[:i][-1]
            pred_last_value.extend(last_value for _ in range(window))
        return pred_last_value
    elif method == 'MA':
        pred_MA = []
        for i in range(train_len, total_len, window):
            arima = ARIMA(arr[:i], order=order)
            arima_fit = arima.fit()
            predictions = arima_fit.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean[-window:]
            pred_MA.extend(oos_pred)
        return pred_MA
    elif method == 'MA2':
        pred_MA = []
        for i in range(train_len, total_len, window):
            arima = ARIMA(arr[:i], order=order)
            arima_fit = arima.fit()
            insample_fitvalue = arima_fit.fittedvalues
            arima_forecast = arima_fit.get_forecast(steps=window)
            oos_pred = arima_forecast.predicted_mean
            pred_MA.extend(oos_pred)
        return pred_MA
```

```python
arima_order = (0, 0, 3)
predicted_mean = rolling_forecast(data_process, len(train), len(test), 2, 'mean', arima_order)
predicted_last_value = rolling_forecast(data_process, len(train), len(test), 2, 'last', arima_order)
predicted_ma = rolling_forecast(data_process, len(train), len(test), 2, 'MA', arima_order)
predicted_ma1 = rolling_forecast(data_process, len(train), len(test), 2, 'MA2', arima_order)
```

```python
fig, ax = plt.subplots(figsize=(20, 6))
ax.plot(test, color='red', label='actual')
ax.plot(predicted_mean, 'b--', label='Mean')
ax.plot(predicted_last_value, color='black', label='Last')
ax.plot(predicted_ma, color='pink', label='MA')
ax.plot(predicted_ma1, color='green', label='My_MA')
ax.legend()

plt.show()
```

## Doing something generic

```python

def rolling_forecast_2(arr: np.ndarray, train_len: int, horizon: int, window: int, method: str, order: Tuple[int, int, int], fitted_func: Callable) -> list:
    total_len = train_len + horizon
    if method == 'mean':
        pred_mean = []
        for i in range(train_len, total_len, window):
            mean = np.mean(arr[:i])
            pred_mean.extend(mean for _ in range(window))
        return pred_mean
    elif method == 'last':
        pred_last_value = []
        for i in range(train_len, total_len, window):
            last_value = arr[:i][-1]
            pred_last_value.extend(last_value for _ in range(window))
        return pred_last_value
    elif method == 'MA':
        pred_MA = []
        for i in range(train_len, total_len, window):
            arima = ARIMA(arr[:i], order=order)
            arima_fit = arima.fit()
            predictions = arima_fit.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean[-window:]
            pred_MA.extend(oos_pred)
        return pred_MA
    elif method == 'MA2':
        pred_MA = []
        for i in range(train_len, total_len, window):
            oos_pred = fitted_func(arr, i, order, window)
            pred_MA.extend(oos_pred)
        return pred_MA
    

def sarimax_func(arr: np.ndarray, i: int, order: Tuple[int, int, int], window: int) -> list:
    model = SARIMAX(arr[:i], order=order)
    res = model.fit(disp=False)
    predictions = res.get_prediction(0, i + window - 1)
    oos_pred = predictions.predicted_mean[-window:]
    return oos_pred


sarimax_order = (0, 0, 3)
predicted_mean1 = rolling_forecast_2(data_process, len(train), len(test), 2, 'mean', arima_order, None)
predicted_last_value1 = rolling_forecast_2(data_process, len(train), len(test), 2, 'last', arima_order, None)
predicted_ma1 = rolling_forecast_2(data_process, len(train), len(test), 2, 'MA', arima_order, sarimax_func)
predicted_myma1 = rolling_forecast_2(data_process, len(train), len(test), 2, 'MA2', arima_order, sarimax_func)
```

```python
mse_mean = mean_squared_error(test, predicted_mean1)
mse_last = mean_squared_error(test, predicted_last_value1)
mse_ma = mean_squared_error(test, predicted_ma1)
mse_myma = mean_squared_error(test, predicted_myma1)

print(mse_mean, mse_last, mse_ma, mse_myma)
```
